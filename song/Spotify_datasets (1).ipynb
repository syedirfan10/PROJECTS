# **Building a Song Recommendation System Using KNN**

```python
def recommend_songs(song_name, df, knn, song_Attributes_1):
    song_index = df[df['track_name'] == song_name].index
    if len(song_index) == 0:
        print(f"Song by '{song_name}' not found in the dataset.")
        return None
    song_index = song_index[0]
    numerical_features = song_Attributes_1.select_dtypes(include=np.number)
    song_vector = numerical_features.iloc[song_index].values.reshape(1, -1)
    distances, indices = knn.kneighbors(song_vector, n_neighbors=6)
    recommended_indices = indices[0][1:]
    recommendations = df.iloc[recommended_indices][['track_name', 'track_artist', 'track_popularity']]

    return recommendations
recommend_songs('Call You Mine - Keanu Silva Remix', df, knn, song_Attributes_1)
```
